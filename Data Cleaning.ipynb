```python
import pandas as pd
```

# 1. Website Sessions

### Import File

```python
ws = pd.read_csv("D:/PROJECTS/Maven+Fuzzy+Factory(E-Commerce)/website_sessions.csv")
ws.head(5)
```

```python
ws.info()
```

### Changing datatype of created_at from object to datetime

```python
ws["created_at"] = pd.to_datetime(ws["created_at"], dayfirst = True)
```

```python
ws.info()
```

### Changing datatype of is_repeat_session from int64 to boolean

```python
ws["is_repeat_session"] = ws["is_repeat_session"].astype("bool")
ws.info()
```

```python
ws["website_session_id"].duplicated().sum()
```

```python
ws.isnull().sum()
```

### Filling NULL values with appropriate values

```python
ws["utm_source"] = ws["utm_source"].fillna("direct") 
ws["utm_campaign"] = ws["utm_campaign"].fillna("none")
ws["utm_content"] = ws["utm_content"].fillna("none")
ws["http_referer"] = ws["http_referer"].fillna("direct")
ws.isnull().sum()
```

```python
ws
```

### Adding year and month, month name column

```python
ws.insert(2, "year", ws["created_at"].dt.year)
ws.insert(3, "month", ws["created_at"].dt.month)
ws.insert(4, "month_name", ws["created_at"].dt.month_name())
ws
```

### Exporting Dataset

```python
ws.to_csv("website_sessions_cleaned.csv", index=False)
```
